# Customer Segmentation for effective allocation of marketing activities

```python
import numpy as np, pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv('Mall_Customers.csv')
```

```python
customers.head()
```

# Exploratory Data Analysis

```python
# Check to see descriptive statistics
customers.describe()
```

1. From calling describe, we can see that there are no values to clean.
2. Age looks pretty normally distributed, annual income in the thousands doesn't have outliers that are too excessive.
3. Spending score is in fact between 1 and 100. Everything looks good.

```python
# See the distribution of gender to recognize different distributions
sns.countplot(x='Genre', data=customers);
plt.title('Distribution of Genre');
```

There are more women than men in this dataset.

```python
# Histogram of ages

customers.hist('Age', bins=35);
plt.title('Distribution of Age');
plt.xlabel('Age');
```

Age histogram is somewhat right-tailed. We saw that the average age was 38 as well, so this is not surprising, with a spike in ages 48-49 and 65 as well.

```python
# Histogram of ages by gender
plt.hist('Age', data=customers[customers['Genre'] == 'Male'], alpha=0.5, label='Male');
plt.hist('Age', data=customers[customers['Genre'] == 'Female'], alpha=0.5, label='Female');
plt.title('Distribution of Age by Genre');
plt.xlabel('Age');
plt.legend();
```

We can see two things here, one reflected earlier in the describe call
1. There are more women than men in this data set.
2. There are a lot of younger women and middle-aged women.

```python
# Histogram of income
customers.hist('Annual Income (k$)');
plt.title('Annual Income Distribution in Thousands of Dollars');
plt.xlabel('Thousands of Dollars');
```

The most frequent annual incomes are between around 50 and 85,000

```python
# Histogram of income by gender
plt.hist('Annual Income (k$)', data=customers[customers['Genre'] == 'Male'], alpha=0.5, label='Male');
plt.hist('Annual Income (k$)', data=customers[customers['Genre'] == 'Female'], alpha=0.5, label='Female');
plt.title('Distribution of Income by Genre');
plt.xlabel('Income (Thousands of Dollars)');
plt.legend();
```

Women generally had lower income than men, the majority falling between 45 and 80,000

```python
# Create data sets by gender
male_customers = customers[customers['Genre'] == 'Male']
female_customers = customers[customers['Genre'] == 'Female']
```

```python
# Print the average spending score for men and women
print(male_customers['Spending Score (1-100)'].mean())
print(female_customers['Spending Score (1-100)'].mean())
```


Women on average had a higher spending score by about 3 points

```python
sns.scatterplot('Age', 'Annual Income (k$)', hue='Genre', data=customers);
plt.title('Age to Income, Colored by Genre');
```

There is pretty much no correlation between age and income for either men or women in this data.
The correlation matrix below confirms this

```python
sns.scatterplot('Age', 'Spending Score (1-100)', hue='Genre', data=customers);
plt.title('Age to Spending Score, Colored by Genre');
```

The above plot shows the negative correlation between age and spending score.
 It's not a strong association, but the older the person, the worse their spending score

```python
sns.heatmap(female_customers.corr(), annot=True);
plt.title('Correlation Heatmap - Female');
```

```python
sns.heatmap(male_customers.corr(), annot=True);
plt.title('Correlation Heatmap - Male');
```

Comparing men to women reveals that there is a slightly higher correlation between age and spending score for women

```python

sns.lmplot('Age', 'Spending Score (1-100)', data=female_customers);
plt.title('Age to Spending Score, Female Only');
```

```python
sns.scatterplot('Annual Income (k$)', 'Spending Score (1-100)', hue='Genre', data=customers);
plt.title('Annual Income to Spending Score, Colored by Genre');
```

# There are 5 rough clusters here:

1. Low income, low spending score
2. Low income, high spending score
3. Mid income, medium spending score
4. High income, low spending score
5. High income, high spending score
6. Interestingly, there are no high income, medium spending score points

# Business Interpretation:
1. Marketing cheaper items to women to see if they purchase more frequently or more volume.

2. Marketing more to younger women because their spending score tends to be higher.

3. TTargeting older women (older than early 40s) who have lower spending scores to convert them into profitable customers.
